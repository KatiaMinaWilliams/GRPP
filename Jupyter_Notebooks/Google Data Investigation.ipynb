```python
def groupSum(l):
    li = list(l)
    seen = np.array([li[0]])
    #summ = li[0]
    curr = li[0]
    for i in range(1, len(li)):
        if li[i] not in seen:
            #summ += "; " + li[i]
            seen = np.append(seen, li[i])
    return list(seen)
        
corp.iloc[:, [0, 1, 2, 3, 7, 9]].groupby("Corporation").agg(groupSum)
```
