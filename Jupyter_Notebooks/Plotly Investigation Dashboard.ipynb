```python
import pandas as pd
import numpy as np
import re
import scipy as sp
import plotly.plotly as py
import plotly.graph_objs as go
```

```python
corp = pd.read_csv("2019.02.21_Cleaned_Corporate_Donor_Sheet.csv").iloc[:, 1:11]
corp.fillna("NaN", inplace = True)
```

```python
#Making the Plotly Table
trace = go.Table(
    header=dict(values=['Corporation', "SDG", "Source Name", "Link", "Explicit/Implicit", "Country"],
                fill = dict(color='#C2D4FF'),
                align = ['left'] * 5),
    cells=dict(values=[corp.Corporation, corp.iloc[:, 1], corp.iloc[:, 2], corp.Source, corp.iloc[:, 7], corp.iloc[:, 9]],
               fill = dict(color='#F5F8FF'),
               align = ['left'] * 5))
py.iplot([trace], filename = 'Corportion SDG Commitment')
```

```python
#Making the Bar chart
corp
countries = np.unique(corp['Relevant Country'])
data = []
for sdg in range(18):
    sub = corp[corp.iloc[:, 1] == sdg]
    pdCountry = pd.DataFrame(np.unique(corp['Relevant Country']))
    pdCountry.set_index(0, inplace = True)
    g = pd.DataFrame(sub.groupby('Relevant Country').agg('count').iloc[:, 0])
    g.merge(pdCountry, 'outer', right_index = True, left_index = True).fillna(0)
    
    trace = go.Bar(
        x = countries,
        y = g.iloc[:, 0],
        name = "SDG " + str(sdg)
    )
    data += trace    
layout = go.Layout(barmode = "stack")
fig = go.Figure(data=data, layout=layout)
py.iplot(fig, filename='stacked-bar')
```
